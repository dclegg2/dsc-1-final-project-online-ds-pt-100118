# Null values in waterfront, view, yr_renovated...how to handle?
- waterfront nulls about 11% of data, yr_renovated nulls 18%, view only 0.3%
    - We could feasibly drop rows with view nulls since we lose very little data, but dropping rows for the other 2 not reasonable
    - Null values don't represent a high enough proportion of the series to drop any of the columns outright
    - Dropping seems like a bad approach (except maybe for view); let's consider replacing with common values instead
    - From the above loop, we can see that all 3 columns contain mostly '0' values and that is clearly the best central measure (mode and median) for each.  It therefore seems logical to replace all "NaN" values with "0".  This seems too easy! Let's unpack the logic a bit:
        - We can safely assume that "NaN" waterfront value means the house is not on the water and should be "0". This is a binary boolean column - '1' meaning 'True' or on the water and '0' meaning 'False' or not on the water.
        - Likewise with view, we can assume "NaN" means that there is no view or a negligible view
        - With yr_renovated, the '0' value implies that the house has never been renovated.  The NaN value likely means the same.  


```python
#replacing all the NaN values with 0.0 as all of these columns are currently stored as floats
for col in ['view', 'waterfront', 'yr_renovated']:
    df[col].fillna(0.0, inplace=True)
```

Let's look at regression plots and qq plots so we can determine whether these variables meet the assumptions for linear regression; in particular, we'll inspect whether there's a linear relationship with price using the scatterplots, the normality of the distribution of residuals (homoskedasticity), and skew/kurtosis using qq plots.

```python
df2 = df3.copy()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
import statsmodels.stats.api as sms
for idx, val in enumerate(['bedbath_ratio','sqft_living','grade_cond_ratio', 'sqft_total15']):
    print ("Regression Analysis and Diagnostics: price and " + val)
    print ("---------------------------------------------------------------------------")
    f = 'price~' + val
    model = smf.ols(formula=f, data=df2).fit()
    X_new = pd.DataFrame({val: [df2[val].min(), df2[val].max()]});
    preds = model.predict(X_new)
    df2.plot(kind='scatter', x=val, y='price');
    plt.plot(X_new, preds, c='red', linewidth=2);
    plt.show()
    fig = plt.figure(figsize=(15,8))
    fig = sm.graphics.plot_regress_exog(model, val, fig=fig)
    fig = sm.graphics.qqplot(model.resid, dist=stats.norm, line='45', fit=True)
    plt.show()
```

From the above it seems clear that these need to be transformed.  There is some obvious heteroskedascity looking at residuals and qqplots show high kurtosis for most of these...in order to proceed and meet our assumptions for linear regression, we'll need to make these more normal and sacrifice some interpretability.  We'll use log transformation for continuous variables, then scale using min-max scaling for our target and z-score standardization for the features.

```python

to_log = ['bedbath_ratio','sqft_living', 'price','grade_cond_ratio','sqft_total15']
df3_log = pd.DataFrame([])
for col in to_log:
    df3_log[col] = np.log(df3[col])
```

```python

df3_log['price'] = (df3_log['price'] - min(df3_log['price']))/(max(df3_log['price']) - min(df3_log['price']))


df3_log_features = df3_log.drop(['price'], axis=1)
for col in df3_log_features.columns:
    df3_log[col] = (df3_log[col] - df3_log[col].mean())/(df3_log[col].std())
```

```python

import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
import statsmodels.stats.api as sms
for idx, val in enumerate(['bedbath_ratio','sqft_living','grade_cond_ratio', 'sqft_total15']):
    print ("Transformed Regression Analysis and Diagnostics: price and " + val)
    print ("---------------------------------------------------------------------------")
    f = 'price~' + val
    model = smf.ols(formula=f, data=df3_log).fit()
    X_new = pd.DataFrame({val: [df3_log[val].min(), df3_log[val].max()]});
    preds = model.predict(X_new)
    df3_log.plot(kind='scatter', x=val, y='price');
    plt.plot(X_new, preds, c='red', linewidth=2);
    plt.show()
    fig = plt.figure(figsize=(15,8))
    fig = sm.graphics.plot_regress_exog(model, val, fig=fig)
    fig = sm.graphics.qqplot(model.resid, dist=stats.norm, line='45', fit=True)
    plt.show();
```

Let's now create a loop to iteratively select features using RFE and inspect the r-squared and adjusted r-squared for each set of features:

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
X = features
y = target
rsq = []
rsq_adj = []
for n_size in list(range(1, 16, 1)):
    linreg = LinearRegression()
    selector = RFE(linreg, n_features_to_select = n_size)
    selector = selector.fit(X, y)
    feature_columns = X.columns[selector.support_]
    linreg.fit(X[feature_columns], y)
    y_hat = linreg.predict(X[feature_columns])
    SSR = np.sum((y-y_hat)**2)
    SST = np.sum((y - np.mean(y))**2)
    r_sq = 1- (SSR/SST)
    r_sq_adj = 1 - (((1-r_sq)*(len(y)-1))/(len(y) - X.shape[1] -1))
    rsq.append(r_sq)
    rsq_adj.append(r_sq_adj)
    print(X.columns[selector.support_])
    print('R-squared for ' + str(n_size) + ' features: ' + str(r_sq))
    print('R-squared-adj for ' + str(n_size) +' features: ' + str(r_sq_adj))
    print('-----------------------------------------------------\n')
```

It seems like the sweet spot is with 10 features, the final addition being view_0...the r-squared jumps from .737 with 9 features to .751 with 10 features, then plateaus after.

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=3)

MSE_train = []
MSE_test = []
for n_size in list(range(1, 16, 1)):
    linreg = LinearRegression()
    selector = RFE(linreg, n_features_to_select = n_size)
    selector = selector.fit(X_train, y_train)
    feature_columns = X.columns[selector.support_]
    linreg.fit(X_train[feature_columns], y_train)
    y_hat_train = linreg.predict(X_train[feature_columns])
    y_hat_test = linreg.predict(X_test[feature_columns])
    mse_train = np.sum((y_train-y_hat_train)**2)/(len(y_train))
    mse_test = np.sum((y_test - y_hat_test)**2)/(len(y_test))
    print("Number of features: ", n_size)
    print("Mean Squared Error of the training set: ", mse_train)    
    print("Mean Squared Error of the testing set: ", mse_test, "\n")
    MSE_train.append(mse_train)
    MSE_test.append(mse_test)
```

We also observe very low MSEs (although that should be expected since we log-normalized our data) and negligible differences between the MSE of the training and testing sets.
